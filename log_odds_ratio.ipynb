```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import seaborn as sns

%matplotlib inline
```

```python
li = []
```

```python
sample_size = 356 #  I didn't want to keep sample size in mind
```

```python
#  Creating 29 "cancer" written cards and putting them in a bag

for i in range(0, 29):
    li.append("cancer")
```

```python
#  Creating 327 "no cancer" written cards and putting them in a bag

for j in range(0, 327):
    li.append("no cancer")
```

```python
len(li) == sample_size  # Checking to see if I have done anything wrong
```

```python
mutated_version = []
wild_version = []
```

```python
for c in range(10000):
    sampl_all = np.random.choice(li, size=len(li)) #  Shuffling the bag (cards)
    sampl_mutated_version = sampl_all[0:140].tolist() #  Taking 140 of them (representing the number of participants with mutated gene)
    mutated_version.append(sampl_mutated_version.count("cancer") / sampl_mutated_version.count("no cancer"))  # and calculating the odds in favor of cancer in the mutated gene group
    sampl_wild_version = sampl_all[140:356].tolist() #  Taking rest of the cards (representing the number of participants in the non-mutated gene group)
    wild_version.append(sampl_wild_version.count("cancer") / sampl_wild_version.count("no cancer")) #  and calculating the odds in favor of cancer in the non-mutated gene group
```

```python
mutated_version_nd = np.array(mutated_version)
wild_version_nd = np.array(wild_version)
```

```python
odds_ratio = mutated_version_nd / wild_version_nd #  Calculating odds ratio
```

```python
log_odds_ratio = np.log(odds_ratio) #  Calculating log(odds ratio)
```

```python
#  Plotting the distribution of log(odds ratio)

fig = plt.figure(figsize=(10, 8))
g = sns.histplot(data=log_odds_ratio, bins=30, color="orange", edgecolor="black")

g.xaxis.set_minor_locator(AutoMinorLocator())

g.tick_params(which="both", width=2)
g.tick_params(which="major", length=7)
g.tick_params(which="minor", length=4)

#  Apparently it can be approximated by normal distribution
```
